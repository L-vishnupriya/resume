# **precision_score:**

```python
from sklearn.metrics import precision_score

# Example: True and predicted entities (binary labels for simplicity)
y_true = [1, 0, 1, 1, 0]   # 1 = entity present, 0 = not an entity
y_pred = [1, 0, 0, 1, 0]

precision = precision_score(y_true, y_pred)

print("Precision:", precision)
```

# **recall_score:**

```python
from sklearn.metrics import recall_score

# Example: True and predicted entities (binary labels for simplicity)
y_true = [1, 0, 1, 1, 0]   # 1 = entity present, 0 = not an entity
y_pred = [1, 0, 0, 1, 0]


recall = recall_score(y_true, y_pred)

print("Recall:", recall)

```

# **f1_score:**

```python
from sklearn.metrics import f1_score

# Example: True and predicted entities (binary labels for simplicity)
y_true = [1, 0, 1, 1, 0]   # 1 = entity present, 0 = not an entity
y_pred = [1, 0, 0, 1, 0]

f1 = f1_score(y_true, y_pred)

print("F1 Score:", f1)
```
