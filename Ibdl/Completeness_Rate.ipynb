# **Completeness Rate:**

```python
expected_terms = [
    "diabetes", "insulin", "blood pressure", "follow-up"
]

transcribed_text = """
Patient has diabetes and blood pressure issues.
Follow-up scheduled next week.
"""

transcribed_text = transcribed_text.lower()

found_terms = [term for term in expected_terms if term in transcribed_text]

completeness_rate = len(found_terms) / len(expected_terms)

print("Completeness Rate:", completeness_rate)
```

**Output meaning:**

1.0 -fully complete

< 1.0 - some information missing

# **Handling Partial Transcriptions:**

```python
# Example transcription output from ASR system
transcription_segments = [
    {"text": "Patient has diabetes", "confidence": 0.92},
    {"text": "and is prescribed", "confidence": 0.88},
    {"text": "", "confidence": 0.10},   # missing / failed segment
    {"text": "insulin daily", "confidence": 0.95}
]

CONFIDENCE_THRESHOLD = 0.75

partial_segments = []

for i, segment in enumerate(transcription_segments):
    if segment["confidence"] < CONFIDENCE_THRESHOLD or segment["text"].strip() == "":
        partial_segments.append(i)

if partial_segments:
    print("Partial transcription detected at segments:", partial_segments)
else:
    print("Transcription is complete")
```

# **Handling Partial Transcriptions:**

```python
import librosa
import numpy as np
import soundfile as sf

# Load audio
audio, sr = librosa.load("audio.wav", sr=16000)

# Normalize volume
audio = audio / np.max(np.abs(audio))

# Remove silence
audio_trimmed, _ = librosa.effects.trim(audio, top_db=20)

# Simple noise reduction (basic)
stft = librosa.stft(audio_trimmed)
magnitude, phase = librosa.magphase(stft)
magnitude_denoised = np.maximum(magnitude - np.mean(magnitude), 0)
audio_denoised = librosa.istft(magnitude_denoised * phase)

# Save processed audio
sf.write("processed_audio.wav", audio_denoised, sr)

print("processed_audio.wav saved successfully")
```
